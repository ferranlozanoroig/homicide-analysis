```python
import json

with open('../data/ne_110m_admin_0_countries.geojson', 'r', encoding='utf-8') as f:
    geo = json.load(f)
    
print(f'Número de países en GeoJSON: {len(geo["features"])}')
print(f'Propiedades disponibles: {list(geo["features"][0]["properties"].keys())}')
```

```python
# Extraer códigos ISO del GeoJSON
geo_codes = [f['properties']['ISO_A3'] for f in geo['features']]

# Países en nuestro dataset que no están en el GeoJSON
df_countries[~df_countries['Code'].isin(geo_codes)]['Entity'].unique()
```

```python
# ¿Qué código tiene Francia en nuestro dataset?
df_countries[df_countries['Entity'] == 'France']['Code'].unique()
```

```python
# ¿Existe FRA en el GeoJSON?
'FRA' in geo_codes
```

```python
# Buscar Francia en el GeoJSON por nombre
for f in geo['features']:
    if 'France' in f['properties']['NAME']:
        print(f['properties']['NAME'], f['properties']['ISO_A3'])
```

```python
for f in geo['features']:
    if f['properties']['ISO_A3'] == '-99':
        print(f['properties']['NAME'], f['properties']['ADM0_A3'])
```

```python
for f in geo['features']:
    if f['properties']['ISO_A3'] == '-99':
        f['properties']['ISO_A3'] = f['properties']['ADM0_A3']

# Verificar que Francia ahora tiene FRA
for f in geo['features']:
    if f['properties']['NAME'] == 'France':
        print(f['properties']['NAME'], f['properties']['ISO_A3'])
```

```python
geo_codes = [f['properties']['ISO_A3'] for f in geo['features']]
sin_mapa = df_countries[~df_countries['Code'].isin(geo_codes)]['Entity'].unique()
print(f'Países sin correspondencia: {len(sin_mapa)}')
print(sin_mapa)
```

```python
import json

with open('../data/countries-clean.geojson', 'w', encoding='utf-8') as f:
    json.dump(geo, f)

print('GeoJSON guardado correctamente')
```
